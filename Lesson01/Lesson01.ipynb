```python
import scipy.stats as stats

# 学生的数学成绩数据
x = [85, 92, 78, 89, 95, 80]

# 学生的语文成绩数据
y = [80, 88, 75, 92, 90, 85]

# 计算Pearson相关系数和p-value
corr, p_value = stats.pearsonr(x, y)

# 打印结果
print("Pearson相关系数:", corr)
print("p-value:", p_value)
```

```python
import scipy.stats as stats

# 学生的数学成绩数据
math_scores = [85, 92, 78, 89, 95, 80]

# 学生的语文成绩数据
chinese_scores = [80, 88, 75, 92, 90, 85]

# 计算Spearman相关系数和p-value
corr, p_value = stats.spearmanr(math_scores, chinese_scores)

# 打印结果
print("Spearman相关系数:", corr)
print("p-value:", p_value)
```

```python
import scipy.stats as stats

# 准备数据
data = [3, 4, 2, 6, 8, 5, 7, 1, 9]  # 样本数据
population_mean = 5  # 总体均值

# 进行单样本t检验
t_statistic, p_value = stats.ttest_1samp(data, population_mean)

# 打印结果
print("t统计量:", t_statistic)
print("p-value:", p_value)

# 结论判断
alpha = 0.05  # 显著性水平
if p_value <= alpha:
    print("拒绝原假设，样本均值与总体均值存在显著差异")
else:
    print("无法拒绝原假设，样本均值与总体均值无显著差异")
```

```python
from scipy.stats import chi2_contingency

# 创建一个包含观察值的2x2的列联表
observed_values = [
    [30, 10],
    [20, 25]
]

# 执行卡方检验
chi2, p, dof, expected = chi2_contingency(observed_values)

# 打印结果
print("卡方统计量: ", chi2)
print("p-value: ", p)
print("自由度: ", dof)
print("期望值: ", expected)
```

```python
from scipy.stats import f_oneway

# 创建三个样本数据
sample1 = [1, 2, 3, 4, 5]
sample2 = [2, 4, 6, 8, 10]
sample3 = [3, 6, 9, 12, 15]

# 执行方差检验
f_statistic, p_value = f_oneway(sample1, sample2, sample3)

# 打印结果
print("F-statistic: ", f_statistic)
print("p-value: ", p_value)
```
